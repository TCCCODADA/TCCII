# Rede Neural Recorrente (RNN) LSTM (Long Short-Term Memory - Memória de Curto e Longo Prazo)
Composta por uma camada LSTM com 50 neurônios e ativação tanh, seguida por uma camada de BatchNormalization para estabilizar o treinamento.

```python
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.models import Sequential
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

features = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume', 'SMA_10', 'EMA_10', 
            'RSI_10', 'MACD', 'MACD_Histogram', 'BBL_20_2.0', 'BBM_20_2.0', 'BBU_20_2.0', 
            'BBB_20_2.0', 'BBP_20_2.0', 'ATRr_14', 'stoch_k', 'stoch_d']  # Define as features que serão usadas para treinar o modelo

target = 'Adj Close'  # Define o alvo que será previsto, neste caso o preço ajustado de fechamento

X = df[features]  # Seleciona as features do DataFrame
y = df[target]  # Seleciona a coluna alvo

scaler = MinMaxScaler()  # Inicializa o scaler para normalizar os dados
X_scaled = scaler.fit_transform(X)  # Normaliza os dados de entrada para um intervalo de 0 a 1

y = y.values  # Converte o alvo em um array NumPy

window_size = 60  # Define o tamanho da janela para a criação de sequências

X_seq = []  # Inicializa a lista para armazenar sequências de entrada
y_seq = []  # Inicializa a lista para armazenar as saídas correspondentes

# Cria sequências de dados de entrada e saída
for i in range(window_size, len(X_scaled)):
    X_seq.append(X_scaled[i-window_size:i])  # Adiciona uma sequência de 60 pontos de dados de entrada
    y_seq.append(y[i])  # Adiciona o valor de saída correspondente

X_seq = np.array(X_seq)  # Converte as sequências de entrada para um array NumPy
y_seq = np.array(y_seq)  # Converte as saídas para um array NumPy

train_size = int(len(X_seq) * 0.7)  # Define o tamanho do conjunto de treino como 70% dos dados
X_train, X_test = X_seq[:train_size], X_seq[train_size:]  # Divide os dados de entrada em treino e teste
y_train, y_test = y_seq[:train_size], y_seq[train_size:]  # Divide os dados de saída em treino e teste

accuracies = []  # Lista para armazenar as acurácias de cada loop
best_accuracy = -np.inf  # Inicializa a melhor acurácia com um valor muito baixo
best_model = None  # Inicializa o melhor modelo
best_y_pred = None  # Inicializa as previsões do melhor modelo

low_accuracy = +np.inf  # Inicializa a pior acurácia com um valor muito alto
low_model = None  # Inicializa o pior modelo
low_y_pred = None  # Inicializa as previsões do pior modelo

for loop in range(5):  # Realiza o treinamento do modelo 5 vezes
    print(f'Treinamento {loop + 1}/5')
    # Cria o modelo sequencial
    model = Sequential()
    model.add(LSTM(50, activation='tanh', input_shape=(X_train.shape[1], X_train.shape[2])))  # Adiciona uma camada LSTM
    model.add(BatchNormalization())  # Adiciona uma camada BatchNormalization para estabilizar o aprendizado
    
    model.add(Dense(50, activation='relu'))  # Adiciona uma camada densa com 50 neurônios e ativacao ReLU
    model.add(Dense(1))  # Adiciona a camada de saída com um neurônio

    # Compila o modelo, usando o otimizador 'adam' e a função de perda 'mse' (erro quadrático médio)
    model.compile(optimizer='adam', loss='mse')

    # Treina o modelo com os dados de treino, usando 50 épocas e um tamanho de batch de 32
    history = model.fit(X_train, y_train, epochs=30, batch_size=32, validation_data=(X_test, y_test))  # Reduz o número de épocas para 30 para evitar overfitting

    # Faz previsões para os dados de teste
    y_pred = model.predict(X_test)  # Faz previsões para os dados de teste

    # Calcula o erro absoluto médio percentual (MAPE)
    mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100  # Calcula o erro percentual médio
    mae = mean_absolute_error(y_test, y_pred)  # Calcula o erro absoluto médio
    r2 = r2_score(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)  # Calcula o erro quadrático médio
    print(f'MAPE: {mape:.2f}%')  # Exibe o MAPE
    print(f'MAE: {mae}')
    print(f'MSE: {mse}')
    print(f'R²: {r2}')

    # Calcula a porcentagem de acurácia usando MAPE
    accuracy = max(0, 100 - mape)  # Garante que a acurácia não seja negativa
    accuracies.append(accuracy)  # Armazena a acurácia deste loop
    print(f'Acurácia: {accuracy:.2f}%')  # Exibe a acurácia em porcentagem

    # Verifica se este é o melhor modelo
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = model
        best_y_pred = y_pred

        best_mape = mape
        best_mae = mae
        best_mse = mse
        best_r2 = r2

    # Verifica se este é o pior modelo
    if accuracy < low_accuracy:
        low_accuracy = accuracy
        low_model = model
        low_y_pred = y_pred

        low_mape = mape
        low_mae = mae
        low_mse = mse
        low_r2 = r2

# Plota as acurácias de cada loop
plt.figure(figsize=(10, 5))
plt.plot(range(1, 6), accuracies, marker='o', linestyle='-', color='b', label='Acurácia por Loop')
plt.title('Acurácia de Cada Loop de Treinamento')
plt.xlabel('Loop de Treinamento')
plt.ylabel('Acurácia (%)')
plt.xticks(range(1, 6))
plt.legend()
plt.show()

# MELHOR

# Calcula e exibe a acurácia do Best
print(f'Melhor MAPE: {best_mape:.2f}%')  # Exibe o MAPE
print(f'Melhor MAE: {best_mae}')
print(f'Melhor MSE: {best_mse}')
print(f'Melhor R²: {best_r2}')

# Calcula e exibe a acurácia média
mean_accuracy = np.mean(accuracies)
print(f'Acurácia Média: {mean_accuracy:.2f}%')
print(f'Melhor Acurácia: {best_accuracy:.2f}%')

# Obtém as datas correspondentes aos dados de teste
test_dates = df['Date'][-len(y_test):]
test_dates = test_dates.reset_index(drop=True)  # Reseta o índice para uma visualização correta

# Plota os valores reais e previstos para o conjunto de teste do melhor modelo
plt.figure(figsize=(14, 5))
plt.plot(test_dates, y_test, color='blue', label='Valor Real')  # Plota o valor real
plt.plot(test_dates, best_y_pred, color='red', label='Previsões (Melhor Modelo)')  # Plota as previsões do melhor modelo
plt.title('Previsão do Preço das Ações (Melhor Loop)')  # Define o título do gráfico
plt.xlabel('Data')  # Define o rótulo do eixo x
plt.ylabel('Preço das Ações')  # Define o rótulo do eixo y
plt.legend()  # Adiciona uma legenda ao gráfico
plt.show()  # Mostra o gráfico


# PIOR

# Calcula e exibe a acurácia do Low
print(f'Pior MAPE: {low_mape:.2f}%')  # Exibe o MAPE
print(f'Pior MAE: {low_mae}')
print(f'Pior MSE: {low_mse}')
print(f'Pior R²: {low_r2}')

# Calcula e exibe a acurácia média
print(f'Acurácia Média: {mean_accuracy:.2f}%')
print(f'Pior Acurácia: {low_accuracy:.2f}%')

# Plota os valores reais e previstos para o conjunto de teste do pior modelo
plt.figure(figsize=(14, 5))
plt.plot(test_dates, y_test, color='blue', label='Valor Real')  # Plota o valor real
plt.plot(test_dates, low_y_pred, color='red', label='Previsões (Pior Modelo)')  # Plota as previsões do pior modelo
plt.title('Previsão do Preço das Ações (Pior Loop)')  # Define o título do gráfico
plt.xlabel('Data')  # Define o rótulo do eixo x
plt.ylabel('Preço das Ações')  # Define o rótulo do eixo y
plt.legend()  # Adiciona uma legenda ao gráfico
plt.show()  # Mostra o gráfico
```

# Rede Neural Convolucional Unidimensional.
Duas camadas convolucionais 1D, cada uma seguida por uma camada de MaxPooling e Dropout.
As camadas convolucionais extraem padrões temporais e as camadas densa e de Flatten processam essas características para fazer previsões. 

```python
from tensorflow.keras.layers import Conv1D, Dense, Flatten, MaxPooling1D, Dropout
from tensorflow.keras.models import Sequential
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Definindo as features que serão usadas para treinar o modelo
features = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume', 'SMA_10', 'EMA_10', 
            'RSI_10', 'MACD', 'MACD_Histogram', 'BBL_20_2.0', 'BBM_20_2.0', 'BBU_20_2.0', 
            'BBB_20_2.0', 'BBP_20_2.0', 'ATRr_14', 'stoch_k', 'stoch_d']

target = 'Adj Close'  # Define o alvo que será previsto, neste caso o preço ajustado de fechamento

# Seleciona as features do DataFrame e a coluna alvo
y = df[target]
X = df[features]

# Inicializa o scaler para normalizar os dados e normaliza os dados de entrada para um intervalo de 0 a 1
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

y = y.values  # Converte o alvo em um array NumPy

# Define o tamanho da janela para a criação de sequências
window_size = 60

X_seq = []  # Inicializa a lista para armazenar sequências de entrada
y_seq = []  # Inicializa a lista para armazenar as saídas correspondentes

# Cria sequências de dados de entrada e saída
for i in range(window_size, len(X_scaled)):
    X_seq.append(X_scaled[i-window_size:i])  # Adiciona uma sequência de 60 pontos de dados de entrada
    y_seq.append(y[i])  # Adiciona o valor de saída correspondente

# Converte as sequências de entrada e saídas para arrays NumPy
X_seq = np.array(X_seq)
y_seq = np.array(y_seq)

# Divide os dados em treino (70%) e teste (30%)
train_size = int(len(X_seq) * 0.7)
X_train, X_test = X_seq[:train_size], X_seq[train_size:]
y_train, y_test = y_seq[:train_size], y_seq[train_size:]

accuracies = []  # Lista para armazenar as acurácias de cada loop
best_accuracy = -np.inf  # Inicializa a melhor acurácia com um valor muito baixo
best_model = None  # Inicializa o melhor modelo
best_y_pred = None  # Inicializa as previsões do melhor modelo

low_accuracy = +np.inf  # Inicializa a pior acurácia com um valor muito alto
low_model = None  # Inicializa o pior modelo
low_y_pred = None  # Inicializa as previsões do pior modelo

for loop in range(5):  # Realiza o treinamento do modelo 5 vezes
    print(f'Treinamento {loop + 1}/5')
    
    # Cria o modelo sequencial
    model = Sequential()
    # Primeira camada convolucional para extrair padrões temporais dos dados sequenciais
    model.add(Conv1D(filters=64, kernel_size=5, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])))
    model.add(MaxPooling1D(pool_size=2))  # Camada de pooling para reduzir a dimensionalidade
    model.add(Dropout(0.2))  # Adiciona Dropout para evitar overfitting
    
    # Segunda camada convolucional para capturar padrões mais complexos
    model.add(Conv1D(filters=32, kernel_size=3, activation='relu'))
    model.add(MaxPooling1D(pool_size=2))  # Outra camada de pooling
    model.add(Dropout(0.2))  # Outro Dropout para evitar overfitting

    model.add(Flatten())  # Achata a saída da camada convolucional para que possa ser usada na camada densa
    model.add(Dense(50, activation='relu'))  # Camada densa com 50 neurônios para processar as características extraídas
    model.add(Dense(1))  # Camada de saída com um único neurônio para prever o valor final

    # Compila o modelo, usando o otimizador 'adam' e a função de perda 'mse' (erro quadrático médio)
    model.compile(optimizer='adam', loss='mse')

    # Treina o modelo com os dados de treino, usando 50 épocas e um tamanho de batch de 32
    history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))

    # Faz previsões para os dados de teste
    y_pred = model.predict(X_test)

    # Calcula o erro absoluto médio percentual (MAPE)
    mape = np.mean(np.abs((y_test - y_pred) / np.maximum(y_test, 1e-10))) * 100  # Calcula o erro percentual médio e evita divisão por zero
    mae = mean_absolute_error(y_test, y_pred)  # Calcula o erro absoluto médio
    mse = mean_squared_error(y_test, y_pred)  # Calcula o erro quadrático médio
    r2 = r2_score(y_test, y_pred)  # Calcula o coeficiente de determinação R²
    
    # Exibe as métricas de avaliação
    print(f'MAPE: {mape:.2f}%')  # Exibe o MAPE
    print(f'MAE: {mae}')  # Exibe o MAE
    print(f'MSE: {mse}')  # Exibe o MSE
    print(f'R²: {r2}')  # Exibe o R²

    # Calcula a porcentagem de acurácia usando MAPE
    accuracy = max(0, 100 - mape)  # Garante que a acurácia não seja negativa
    accuracies.append(accuracy)  # Armazena a acurácia deste loop
    print(f'Acurácia: {accuracy:.2f}%')  # Exibe a acurácia em porcentagem

    # Verifica se este é o melhor modelo
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = model
        best_y_pred = y_pred

        best_mape = mape
        best_mae = mae
        best_mse = mse
        best_r2 = r2

    # Verifica se este é o pior modelo
    if accuracy < low_accuracy:
        low_accuracy = accuracy
        low_model = model
        low_y_pred = y_pred

        low_mape = mape
        low_mae = mae
        low_mse = mse
        low_r2 = r2

# Plota as acurácias de cada loop
plt.figure(figsize=(10, 5))
plt.plot(range(1, 6), accuracies, marker='o', linestyle='-', color='b', label='Acurácia por Loop')
plt.title('Acurácia de Cada Loop de Treinamento')
plt.xlabel('Loop de Treinamento')
plt.ylabel('Acurácia (%)')
plt.xticks(range(1, 6))
plt.legend()
plt.show()

# MELHOR

# Calcula e exibe a acurácia do Best
print(f'Melhor MAPE: {best_mape:.2f}%')  # Exibe o MAPE
print(f'Melhor MAE: {best_mae}')
print(f'Melhor MSE: {best_mse}')
print(f'Melhor R²: {best_r2}')

# Calcula e exibe a acurácia média
mean_accuracy = np.mean(accuracies)
print(f'Acurácia Média: {mean_accuracy:.2f}%')
print(f'Melhor Acurácia: {best_accuracy:.2f}%')

# Obtém as datas correspondentes aos dados de teste
test_dates = df['Date'][-len(y_test):]
test_dates = test_dates.reset_index(drop=True)  # Reseta o índice para uma visualização correta

# Plota os valores reais e previstos para o conjunto de teste do melhor modelo
plt.figure(figsize=(14, 5))
plt.plot(test_dates, y_test, color='blue', label='Valor Real')  # Plota o valor real
plt.plot(test_dates, best_y_pred, color='red', label='Previsões (Melhor Modelo)')  # Plota as previsões do melhor modelo
plt.title('Previsão do Preço das Ações (Melhor Loop)')
plt.xlabel('Data')  # Define o rótulo do eixo x
plt.ylabel('Preço das Ações')  # Define o rótulo do eixo y
plt.legend()  # Adiciona uma legenda ao gráfico
plt.show()  # Mostra o gráfico


# PIOR

# Calcula e exibe a acurácia do Low
print(f'Pior MAPE: {low_mape:.2f}%')  # Exibe o MAPE
print(f'Pior MAE: {low_mae}')
print(f'Pior MSE: {low_mse}')
print(f'Pior R²: {low_r2}')

# Calcula e exibe a acurácia média e a pior acurácia
print(f'Acurácia Média: {mean_accuracy:.2f}%')
print(f'Pior Acurácia: {low_accuracy:.2f}%')

# Plota os valores reais e previstos para o conjunto de teste do pior modelo
plt.figure(figsize=(14, 5))
plt.plot(test_dates, y_test, color='blue', label='Valor Real')  # Plota o valor real
plt.plot(test_dates, low_y_pred, color='red', label='Previsões (Pior Modelo)')  # Plota as previsões do pior modelo
plt.title('Previsão do Preço das Ações (Pior Loop)')  # Define o título do gráfico
plt.xlabel('Data')  # Define o rótulo do eixo x
plt.ylabel('Preço das Ações')  # Define o rótulo do eixo y
plt.legend()  # Adiciona uma legenda ao gráfico
plt.show()  # Mostra o gráfico
```

```python
years = df['Date'].dt.year.unique()

mses = []
maes = []
mses_geral = []
maes_geral = []
std_mses_geral =[]
std_maes_geral = []
for i in range (10): 
    
    for year in years[:-1]:  # Excluindo o último ano para ter dados para testar
        # Dados de treinamento até o ano atual
        train_data = df[df['Date'].dt.year <= year]
        test_data = df[df['Date'].dt.year == year+1]
        
        if len(test_data) < window_size:
            continue  # Pula se não houver dados suficientes para testar
        
        # Preparar dados de treinamento
        X_train = train_data[features].values
        y_train = train_data[target].values
        X_train_scaled = scaler.fit_transform(X_train)
        
        X_train_seq = []
        y_train_seq = []
        for i in range(window_size, len(X_train_scaled)):
            X_train_seq.append(X_train_scaled[i-window_size:i])
            y_train_seq.append(y_train[i])
        X_train_seq = np.array(X_train_seq)
        y_train_seq = np.array(y_train_seq)
        
        # Preparar dados de teste
        X_test = test_data[features].values
        y_test = test_data[target].values
        X_test_scaled = scaler.transform(X_test)
        
        X_test_seq = []
        y_test_seq = []
        for i in range(window_size, len(X_test_scaled)):
            X_test_seq.append(X_test_scaled[i-window_size:i])
            y_test_seq.append(y_test[i])
        X_test_seq = np.array(X_test_seq)
        y_test_seq = np.array(y_test_seq)
        
        # Verificar se há dados suficientes
        if X_train_seq.shape[0] == 0 or X_test_seq.shape[0] == 0:
            continue
        
        # Construir e treinar o modelo
        model = Sequential()
        model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(X_train_seq.shape[1], X_train_seq.shape[2])))
        model.add(Flatten())
        model.add(Dense(50, activation='relu'))
        model.add(Dense(1))
        model.compile(optimizer='adam', loss='mse')
        model.fit(X_train_seq, y_train_seq, epochs=20, batch_size=32, verbose=0)
        
        # Fazer previsões
        y_pred = model.predict(X_test_seq)
        
        # Calcular métricas
        mse = mean_squared_error(y_test_seq, y_pred)
        mae = mean_absolute_error(y_test_seq, y_pred)
        mses.append(mse)
        maes.append(mae)
        
        print(f'Ano {year+1}: MSE = {mse}, MAE = {mae}')

    mean_mse = np.mean(mses)
    std_mse = np.std(mses)
    mean_mae = np.mean(maes)
    std_mae = np.std(maes)
    mses_geral.append(mean_mse)
    std_mses_geral.append(std_mse)
    maes_geral.append(mean_mae)
    std_maes_geral.append(std_mae)

mean_mse_final = np.mean(mses_geral)
std_mse_final = np.std(std_mses_geral)
mean_mae_final = np.mean(maes_geral)
std_mae_final = np.std(std_maes_geral)
print(f'\nMédia MSE: {mean_mse_final}, Desvio Padrão MSE: {std_mse_final}')
print(f'Média MAE: {mean_mae_final}, Desvio Padrão MAE: {std_mae_final}')

```
